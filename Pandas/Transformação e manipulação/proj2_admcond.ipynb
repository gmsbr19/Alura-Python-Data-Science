# Aula 1

```python
import pandas as pd
```

```python
dados = pd.read_json('dados_locacao_imoveis.json')
dados = pd.json_normalize(dados['dados_locacao'])
dados
```

# Aula 2

```python
# Remover os dados em listas dentro do DataFrame;
cols_a_explode = list(dados.columns)[1:]
cols_a_explode
```

```python
dados = dados.explode(cols_a_explode)
dados.reset_index(drop=True, inplace=True)
dados
```

```python
# Verificar os tipos de dados;
dados.dtypes
```

```python
# Identificar colunas numéricas;
cols_nums = dados['valor_aluguel']
cols_nums
```

```python
# Transformar a coluna numérica para o tipo numérico.
import numpy as np
dados['valor_aluguel'] = cols_nums.apply(lambda x: x.replace('$', '').replace(',', '.').replace('reais', '').strip()).astype(np.float64)
dados.dtypes
```

# Aula 3
Manipule os textos na coluna apartamento para remover o texto (blocoAP) do DataFrame.

```python
dados['apartamento'] = dados['apartamento'].str.replace(' (blocoAP)', '')
```

```python
dados
```

# Aula 4
Na coluna Data de venda, temos datas em formato 'dia/mês/ano' (dd/mm/AAAA). Transforme esses dados para o tipo datetime e busque uma forma de visualização de subconjunto que possa contribuir no objetivo do contexto que os dados estão inseridos.

```python
colunas_datas = dados.columns[1:3]
colunas_datas
```

```python
dados[colunas_datas] = dados[colunas_datas].apply(pd.to_datetime, format='%d/%m/%Y')
dados
```

```python
dados['dias_atraso_pagamento'] = (dados['datas_de_pagamento'] - dados['datas_combinadas_pagamento']).dt.days
dados.groupby('apartamento')[['dias_atraso_pagamento']].mean()
```
