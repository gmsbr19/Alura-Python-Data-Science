# Aula 1

```python
import pandas as pd
```

```python
dados = pd.read_json('dados_vendas_clientes.json')
dados = pd.json_normalize(dados['dados_vendas'])
dados
```

# Aula 2

```python
# Remover os dados em listas dentro do DataFrame;
colunas_a_explode = list(dados.columns)
colunas_a_explode
```

```python
dados = dados.explode(colunas_a_explode[1:])
dados.reset_index(drop=True, inplace=True)
```

```python
# Verificar os tipos de dados;
dados.dtypes
dados
```

```python
# Identificar colunas numéricas;
cols_nums = dados['Valor da compra']
```

```python
# Transformar a coluna numérica para o tipo numérico.
import numpy as np
dados['Valor da compra'] = cols_nums.apply(lambda x: x.replace('R$', '').replace(',', '.').strip()).astype(np.float64)
dados.dtypes
dados
```

# Aula 3
Manipule os textos presentes na coluna Cliente para que seja obtido como resultado os nomes dos clientes em letras minúsculas, com a ausência de caracteres especiais ou números.

```python
dados['Cliente'] = dados['Cliente'].str.lower().replace('(?:[\@|\_])|(?:\s[0-9]+)', '', regex=True)
```

```python
dados
```

# Aula 4
Na coluna Data de venda, temos datas em formato 'dia/mês/ano' (dd/mm/AAAA). Transforme esses dados para o tipo datetime e busque uma forma de visualização de subconjunto que possa contribuir no objetivo do contexto que os dados estão inseridos.

```python
dados['Data de venda']
dias_da_sem = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sab', 'Dom']
dados['Data de venda'] = pd.to_datetime(dados['Data de venda'], format='%d/%m/%Y').dt.day_of_week.apply(lambda x: dias_da_sem[x])
```

```python
dados['Data de venda'] = pd.Categorical(dados['Data de venda'], dias_da_sem[:5])
```

```python
clientes_por_dia = dados.groupby(['Data de venda', 'Cliente']).sum().sort_values('Data de venda')
clientes_por_dia
```
